### Plotting the results from the hek Cycle 24 flare search

```python
import numpy as np
import matplotlib.pyplot as plt
from sunpy.timeseries import TimeSeries
from sunpy.time import TimeRange, parse_time
from sunpy.net import hek
from sunpy.instr import goes

%matplotlib inline
```

```python
# Just some plotting setup - need stix installed 
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams["figure.figsize"] = (16,6)
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'STIXGeneral'

# Some colours for the different GOES Classes
ctc='teal'
ctm='rebeccapurple'
ctx='coral'
ctcx='black'
```

```python
# Load in each of the different search results 
# and plot GOES Flux, Latitude and Number of flare vs time,
# 
fl_src=['SWPC','SSW Latest Events','GOES']
for f in fl_src:
    fl = np.load('fl_'+f.replace(" ","")+'.npz')
    print(f)
    cx_ind=np.argwhere((fl['goes_flux'] >= 1e-6)).flatten()
    c_ind=np.argwhere((fl['goes_flux'] >= 1e-6) & (fl['goes_flux'] < 1e-5)).flatten()
    m_ind=np.argwhere((fl['goes_flux'] >= 1e-5) & (fl['goes_flux'] < 1e-4)).flatten()
    x_ind=np.argwhere(fl['goes_flux'] > 1e-4)
    print('Number of >C: ',len(cx_ind))
    print('Number of C-Class: ',len(c_ind))
    print('Number of M-Class: ',len(m_ind))
    print('Number of X-Class: ',len(x_ind))

    # ----------------------------
    fig,ax = plt.subplots()
    ax.scatter(fl['peak_time'][c_ind],fl['goes_flux'][c_ind],\
               color=ctc,marker='.',s=50,edgecolors='none',alpha=0.5)
    ax.scatter(fl['peak_time'][m_ind],fl['goes_flux'][m_ind],\
               color=ctm,marker='.',s=100,edgecolors='none',alpha=0.6)
    ax.scatter(fl['peak_time'][x_ind],fl['goes_flux'][x_ind],\
               color=ctx,marker='.',s=200,edgecolors='none',alpha=0.7)
    ax.set_xlabel('Year')
    ax.set_ylabel('GOES Flux [Wm$^{-2}$]')
    ax.set_yscale('log')
    ax.set_xlim([2007,2019.9])
    ax.set_ylim([5e-7,2e-3])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
    plt.savefig('goes_flux'+f.replace(" ","")+'.pdf',dpi=150,bbox_inches='tight',pad_inches=0.5, transparent=True)
    plt.show()

    # ----------------------------
    fig,ax = plt.subplots()
    ax.scatter(fl['peak_time'][c_ind],fl['lat'][c_ind],color=ctc,marker='$\mathrm{C}$',s=100,\
               label=str(len(c_ind)),edgecolors='none')
    ax.scatter(fl['peak_time'][m_ind],fl['lat'][m_ind],color=ctm,marker='$\mathrm{M}$',s=200,\
               label=str(len(m_ind)),edgecolors='none')
    ax.scatter(fl['peak_time'][x_ind],fl['lat'][x_ind],color=ctx,marker='$\mathrm{X}$',s=300,\
               label=str(len(x_ind)),edgecolors='none')
    ax.plot([2007,2019.9],[0,0],lw=1,color='grey',ls='dashed')
    ax.set_xlabel('Year')
    ax.set_ylabel('Latitude')
    ax.set_xlim([2007,2019.9])
    ax.set_ylim([-60,60])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.xaxis.label.set_color('k')
    ax.spines['bottom'].set_color('white')
    ax.spines['left'].set_color('white')
    ax.tick_params(axis="x", length=0, width=0)
    ax.tick_params(axis="y", length=0, width=0)
    ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
    plt.legend(fontsize='small',title='Total: '+str(len(cx_ind)))
    plt.savefig('lat'+f.replace(" ","")+'.pdf',dpi=150,bbox_inches='tight',pad_inches=0.5, transparent=True)
    plt.show()

    # ----------------------------
    histc, bins = np.histogram(fl['peak_time'][c_ind], bins=np.arange(2008.5, 2019.5,1/12.))
    histm, bins = np.histogram(fl['peak_time'][m_ind], bins=np.arange(2008.5, 2019.5,1/12.))
    histx, bins = np.histogram(fl['peak_time'][x_ind], bins=np.arange(2008.5, 2019.5,1/12.))
    histcx, bins = np.histogram(fl['peak_time'][cx_ind], bins=np.arange(2008.5, 2019.5,1/12.))
    x = (bins[:-1] + bins[1:]) / 2
    dx=x[1]-x[0]

    fig,ax = plt.subplots()
    # ax.bar(x,histcx,dx,color=ctcx,label='C-Class: '+str(len(cx_ind)))
    ax.bar(x,histc,dx,color=ctc,label='C: '+str(len(c_ind)))
    ax.bar(x,histm,dx,color=ctm,label='M: '+str(len(m_ind)))
    ax.bar(x,histx,dx,color=ctx,label='X: '+str(len(x_ind)))
    ax.set_xlabel('Year')
    ax.set_ylabel('Monthly Number of Flares')
    ax.set_xlim([2007,2019.9])
    ax.set_ylim([0.9,5e2])
    ax.set_yscale('log')
    ax.set_yticks([1,10,100])
    ax.set_yticklabels(['1','10','100'])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
    plt.legend(fontsize='small',loc=2,title='Total: '+str(len(cx_ind)))
    plt.savefig('flarenum'+f.replace(" ","")+'.pdf',dpi=150,bbox_inches='tight',pad_inches=0.5, transparent=True)
    plt.show()
```
