### Plot Cycle 24 Sunspots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Just some plotting setup - need stix installed 
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'STIXGeneral'
```

```python
# from http://www.sidc.be/silso/datafiles
ssn = pd.read_csv('SN_m_tot_V2.0.csv',sep=';',header=None)
```

```python
# Make a nice plot
fig,ax = plt.subplots(figsize=(16,6))
ax.step(ssn[2], ssn[3],lw=3,color='midnightblue')
ax.set_xlabel('Year')
ax.set_ylabel('Monthly <Sunspot Num>')
# ax.set_title('Cycle 24')
ax.set_xlim([2007,2019.9])
ax.set_ylim([0.5,160])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.savefig('ssn.pdf',dpi=150,bbox_inches='tight',pad_inches=0.5,transparent=True)
plt.show()
```
